```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
# we create a csv with the data and the country column is parsed as index in the pandas dataframe
df = pd.read_csv('data.csv', index_col="Country")
```

```python
# here we are slicing the data into features which is X and target variable y
X = df.values[:,:-1]
y = df.values[:,-1]
```

```python
#Here I am plotting the heatmap to find the correlations between the target variable and the features so that it will give is an estimate of how good the fit will be for a linear Regression algorithm
sns.heatmap(data = df.corr(), annot=True)
```

```python
# if the pairplot shows no linearity among the variables than the features are not a good estimation for the target variable if we apply linear regression
# the pairplot is considered to be a good EDA tool for data science as it gives us the trend of how parameters are changing with respect to each other
sns.pairplot(df)
```

```python
#here I am scaling all the features of data to have zero mean and std deviation of 1. So that they are in the same numerical range and a convex function is created which will be easier to optimize
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
# here I am initializing the Kfold and LinearRegression objects with their respective parameters. the random_seed is set for the sake of repeatability of the experiment.
kf = KFold(5, shuffle=True, random_state=42)
lr = LinearRegression()
```

```python
avg_error = 0

for train, test in kf.split(X):#kf.split will give us a generator which will create the dataset with 5 fold
    lr.fit(X[train], y[train]) # we fit the linear Regression object to the training data
    y_preds = lr.predict(X[test]) # we then genreate the predictions of the test data
    mse = mean_squared_error(y[test], y_preds) # then we use the ground truth and the predicted values to find the means squared error for our predictions and the actual values of test data
    fig = plt.figure() # we initialize a figure to plot the ground truth data and the prediction value line for visual inspection
    ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # we add the axes that is required in the figure canvas
    ax.scatter(range(len(y[test])), y[test]) # we create a scatter plot of the ground truth test data
    ax.plot(range(len(y_preds)), y_preds) # we draw a line of the predicted values
    ax.set_title('Ground Truth (Scatter) vs Predicted values (Line)') # adding a title to the plot
    ax.set_xticks(range(len(y_preds)))# we set the ticks values to our x dataset to 
    plt.show() # shows the plots
    avg_error += mse # here we are creating a sum of mean squared error for each fold
avg_error /= 5 # here we are diving the sum of mean squared errors by the 'k' value to get the average error
print(avg_error , 'is the average error calulated over all the folds of data created by the Kfold technique.')
```
